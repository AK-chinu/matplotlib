# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is used for:

Data Visualization: Matplotlib is primarily used for visualizing data in various formats such as line plots, scatter plots, histograms, bar charts, etc.

Publication Quality Graphics: It enables the creation of publication-quality graphics with fine-tuned control over every aspect of the figure.

Integration: It can be integrated into various GUI toolkits such as Tkinter, PyQt, and wxPython, allowing visualization in desktop applications.

Customizability: Matplotlib offers extensive customization options to adjust the appearance of plots according to specific needs.

Interactivity: While Matplotlib primarily produces static plots, it can also create interactive plots when combined with other libraries like Bokeh and Plotly.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot: A basic plot to represent data points connected by straight line segments.

Scatter Plot: Plots individual data points on a two-dimensional plane.

Histogram: Used to represent the distribution of data by dividing it into intervals (bins) and showing how many values fall into each interval.

Bar Chart: A chart that presents categorical data with rectangular bars with heights or lengths proportional to the values they represent.

Pie Chart: A circular statistical graphic divided into slices to illustrate numerical proportions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting scatter plot
plt.scatter(x, y)

# Adding title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plotting vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.title('Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
